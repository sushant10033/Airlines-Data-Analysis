# Q.  number of tickets booked with total amount 

```python
tickets=pd.read_sql_query("""
    SELECT * FROM tickets 
    INNER JOIN bookings 
    ON tickets.book_ref = bookings.book_ref
""", connection)
tickets['book_date']= pd.to_datetime(tickets['book_date'])
tickets['date']=tickets['book_date'].dt.date
tickets
```

```python
x=tickets.groupby('date')[['date']].count()
```

 # Total number of tickets booked per day though line chart

```python
x=tickets.groupby('date')[['date']].count()
plt.figure(figsize=(18,8))
plt.plot(x.index, x['date'],marker='^')
plt.xlabel('Date',fontsize=20)
plt.ylabel('Number of Tickets',fontsize=20)
plt.grid('b')
plt.show()
```

 # sum of amount per day-line chart

```python
bookings=pd.read_sql_query("select * from bookings", connection)
bookings['book_date'] = pd.to_datetime(bookings['book_date'])
bookings['date']=bookings['book_date'].dt.date
x= bookings.groupby('date')[['total_amount']].sum()
plt.figure(figsize=(18,8))
plt.plot(x.index,x['total_amount'],marker='^')
plt.xlabel('Date',fontsize=20)
plt.ylabel('Total amount earned',fontsize=20)
plt.grid('b')
plt.show()
```

# Average charges for each aircraft with different fare condition

```python


df = pd.read_sql_query("""
    SELECT fare_conditions, aircraft_code, AVG(amount)
    FROM ticket_flights 
    JOIN flights ON ticket_flights.flight_id = flights.flight_id 
    GROUP BY aircraft_code, fare_conditions;
""", connection)

df

```

```python
sns.barplot(data=df,x='aircraft_code',y='AVG(amount)',hue='fare_conditions')
```

# Total revenue per year and average revenue per ticket for each aircraft

```python
df = pd.read_sql_query("""
    SELECT aircraft_code, 
           ticket_count, 
           total_revenue, 
           total_revenue / NULLIF(ticket_count, 0) AS avg_revenue_per_ticket 
    FROM (
        SELECT flights.aircraft_code, 
               COUNT(*) AS ticket_count, 
               SUM(ticket_flights.amount) AS total_revenue 
        FROM ticket_flights
        JOIN flights ON ticket_flights.flight_id = flights.flight_id 
        GROUP BY flights.aircraft_code
    ) AS subquery;
""", connection)

df
```

# Average occcupany of seat for each aircraft 

```python
df = pd.read_sql_query("""
    SELECT 
        a.aircraft_code,
        AVG(a.seats_count) AS booked_seats,
        b.num_seats,
        AVG(a.seats_count * 1.0 / b.num_seats) AS occupancy_rate
    FROM (
        SELECT 
            flights.aircraft_code, 
            flights.flight_id, 
            COUNT(*) AS seats_count
        FROM boarding_passes
        INNER JOIN flights 
            ON boarding_passes.flight_id = flights.flight_id
        GROUP BY flights.aircraft_code, flights.flight_id
    ) AS a
    INNER JOIN (
        SELECT 
            aircraft_code, 
            COUNT(*) AS num_seats 
        FROM seats 
        GROUP BY aircraft_code
    ) AS b 
    ON a.aircraft_code = b.aircraft_code 
    GROUP BY a.aircraft_code;
""", connection)

print(df)
```

```python
fig, ax1 = plt.subplots(figsize=(9, 6))


sns.barplot(data=df, x="aircraft_code", y="booked_seats", color="blue", ax=ax1, label="Booked Seats")

ax2 = ax1.twinx()
sns.lineplot(data=df, x="aircraft_code", y="occupancy_rate", marker="o", color="green", linewidth=2, ax=ax2, label="Occupancy Rate")

ax1.set_xlabel("Aircraft Code")
ax1.set_ylabel("Booked Seats", color="blue")
ax2.set_ylabel("Occupancy Rate (%)", color="green")

ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.title("Seats vs. Occupancy Rate per Aircraft")
plt.xticks(rotation=45)
plt.show()
```
